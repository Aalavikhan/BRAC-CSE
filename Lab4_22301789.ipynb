```python
class Patient:
  def __init__(self,id,name,age,blood,prev=None ,next=None):
    self.id=id
    self.name=name
    self.age=age
    self.bloodgroup=blood

class WRM:
  def __init__(self):
    self.head=Patient(None,None,None,None,None)
    self.head.next=self.head
    self.head.prev=self.head
    self.tail=self.head

  def RegisterPatient(self):
    newp=Patient(id=int(input("enter id: ")),name=input("enter name: "),age=int(input("Enter age: ")),blood=input("Enter Blood Group: "))
    newp.next=self.head
    self.head.prev=newp
    newp.prev=self.tail
    self.tail.next=newp
    self.tail=newp
    print("successfully registered\n")
  def ShowAllPatient(self):
    if self.head.next==self.head:
      print("Empty")
    else:
      temp=self.head.next
      while temp.name!=None:
        print(temp.name, end=" ")
        temp=temp.next
      print("\n")
  def ServePatient(self):
    print(f"{self.head.next.name} is served...\n")
    self.head.next=self.head.next.next
    self.head.next.next.prev=self.head
  def CancelAll(self):
    self.head.next=self.head
    self.head.prev=self.head
  def CanDoctorGoHome(self):
    if self.head.next ==self.head:
      print("Yes")
    if self.head.next!=self.head:
      print("No")
  def ReverseTheLine(self):
    current=self.head
    while True:
      # current=self.head
      pr=current.prev
      nx=current.next
      current.next=pr
      current.prev=nx
      current=current.prev
      if current==self.head:
        break
dhdcl=WRM()

while True:

  x= int(input("please select:\n1.Register Patient\n2.Serve Patient\n3.Show All Patient\n4.Cancel all\n5.Can Doctor Go home \n6.Reverse \n7.Exit\n"))
  if x==1:
    dhdcl.RegisterPatient()
  if x==2:
    dhdcl.ServePatient()
  if x==3:
    dhdcl.ShowAllPatient()
  if x==4:
    dhdcl.CancelAll()
  if x==5:
    dhdcl.CanDoctorGoHome()
  if x==6:
    dhdcl.ReverseTheLine()
  if x==7:
    break



# dhdcl=WRM()
# dhdcl.RegisterPatient()
# dhdcl.RegisterPatient()
# dhdcl.RegisterPatient()
# # dhdcl.RegisterPatient()
# dhdcl.ShowAllPatient()
# # dhdcl.CanDoctorGoHome()
# # dhdcl.ServePatient()
# # dhdcl.ServePatient()
# # dhdcl.ShowAllPatient()
# # dhdcl.CancelAll()
# # dhdcl.ShowAllPatient()
# # dhdcl.CanDoctorGoHome()
# dhdcl.ReverseTheLine()
# dhdcl.ShowAllPatient()
```
